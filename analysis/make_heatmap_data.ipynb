```python
# Make summary dataframe containing serum escape and cell entry effects

import pandas as pd

escape_df = pd.read_csv('../results/summaries/Phenotypes_per_antibody_escape.csv').drop(columns='antibody_set')
escape_df.head()
```

```python
phenotypes_df = pd.read_csv('../results/summaries/Phenotypes.csv').drop(columns=['sera escape'])
phenotypes_df.head()
```

```python
# add wildtype rows so can be labeled with 'X' can be included in heatmaps
sera_list = escape_df['antibody'].drop_duplicates().tolist()

wt_df = pd.concat(
    [phenotypes_df.query('wildtype == mutant').assign(antibody=sera) for sera in sera_list],
    ignore_index=True
)[['site', 'wildtype', 'mutant', 'antibody']].assign(escape=0)

escape_df_with_wt = pd.concat(
    [escape_df, wt_df],
    ignore_index=True
)

escape_df_with_wt.head()
```

```python
phenotypes_combined = pd.merge(
    escape_df_with_wt,
    phenotypes_df,
    on=['site', 'wildtype', 'mutant'],
    how='left'
)

phenotypes_combined.to_csv('data/heatmap_data/Phenotypes_per_antibody_escape_with_effects.csv', index=False)
```
